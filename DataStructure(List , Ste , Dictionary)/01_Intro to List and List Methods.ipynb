## <b>Introduction to Lists</b>
Lists are ordered, mutable (changeable) collections of items.

### <b>Creating a List:</b>

```python
number = [2 , 4 , 6 , 8 , 10]
mixed = [1 , "Himani" , True , 3.14]
```

### <b> Common List Methods</b>

```python
my_list = [42, 7, 19, 83, 5, 66, 12, 94, 28, 51]
print("*====APPEND=====*")
print("Before Appending: ",my_list)
my_list.append(86)
print("After Appending: ",my_list)
print("*====INSERT=====*")
print("Before Inserting: ",my_list)
my_list.insert(1,55)
print("After Inserting: ",my_list)
print("*====REMOVE=====*")
print("Before Removing: ",my_list)
my_list.remove(83)
print("After Removing: ",my_list)
print("*====POP=====*")
print("Before Poping: ",my_list)
my_list.pop()
print("After Poping",my_list)
print("*====REVERSING=====*")
print("Before Reversing: ",my_list)
my_list.reverse()
print("After Reversing: ",my_list)
print("*====SORTING=====*")
print("Before Sorting: ",my_list)
my_list.sort()
print("After Sorting: ",my_list)
```

### <b>List Comprehensions (Efficient List Creation)</b>

```python
cube = [x**3 for x  in range(5)]
print(cube)
```
