## <b>Dictionaries and Dictionary Methods</b>
Dictionaries store key-value pairs and allow fast lookups.

### <b>Creating a Dictionary:</b>

```python
students = {"name" : " Himani " , "age": 19 , "grade" : "A" }
```

```python
students
```

### <b>Accessing & Modifying Values:</b>

```python
print(students["name"])  
students["age"] = 22     # Updating value
students["city"] = "Mumbai"  # Adding new key-value pair
```

### <b>Common Dictionary Methods:</b>

```python
print(students.keys())
print(students.values())
print(students.items())

students.pop("age")
students.clear()
```

### <b>When to Use Each Data Structure?</b>
| Data Structure | Features             | Best For                               |
|----------------|----------------------|-----------------------------------------|
| List           | Ordered, Mutable     | Storing sequences, dynamic data         |
| Tuple          | Ordered, Immutable   | Fixed collections, dictionary keys      |
| Set            | Unordered, Unique    | Removing duplicates, set operations     |
| Dictionary     | Key-Value Pairs      | Fast lookups, structured data           |
