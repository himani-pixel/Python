## <b>Sets and Set Methods</b>
Sets are <b>unordered</b>, <b>unique collections</b> (no duplicates).

### <b>Creating a Set:</b>

```python
fruits  = {"Cheery , Rasberry ,  Guava , Pineapple , Orange"}
```

### <b>Key Set Methods:</b>

```python
my_set = {12, 47, 5, 89, 23, 56, 14, 72}
print("*=====ADDING=====*")
print("Before Adding: ",my_set)
my_set.add(52)
print("After Adding: ",my_set)
print("*=====REMOVING=====*")
print("Before Removing: ",my_set)
my_set.remove(5)
print("After Removing: ",my_set)
print("*=====DISCARD=====*")
print("Before Discarding: ",my_set)
my_set.discard(72)  # No error with element is not found
print("After Discarding: ",my_set)
print("*=====POPING=====*")
print("Before Popping: ",my_set)
my_set.pop()
print("After Popping: ",my_set)
```

### <b>Set Operations:</b>

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))      
print(a.intersection(b))  
print(a.difference(b))  
```

<b>Use Case</b>: Sets are great for eliminating duplicate values.
