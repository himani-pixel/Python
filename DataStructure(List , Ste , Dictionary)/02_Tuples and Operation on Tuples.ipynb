## <b>Tuples and Operations on Tuples</b>
Tuples are ordered but immutable collections (cannot be changed after creation).

### <b>Creating a Tuple:</b>

```python
my_tuple = (10 , 20 , 30)
singleElementTuple = (50,) #Comma is imp for single element tuple
```

### <b>Accessing Tuple Elements:</b>

```python
print(my_tuple[1])
```

### <b>Tuple Unpacking:</b>

```python
a , b , c = my_tuple
print(a , b , c)
```

### <b>Common Tuple Methods:</b>

```python
my_tuple1 = (5, 7, 5, 9, 7, 5, 12, 7, 9, 5)
my_tuple1.count(5)
```

```python
my_tuple1.index(12)
```

### <b>Why Use Tuples?</b>
- <b>Faster than lists</b> (since they are immutable)
- <b>Used as dictionary keys</b> (since they are hashable)
- <b>Safe from unintended modifications</b>
