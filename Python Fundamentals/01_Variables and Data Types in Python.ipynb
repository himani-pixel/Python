# Variables and Data Types in Python
## What are Variables?
- Variables are used to store data that can be used and manipulated in a program.
- A variable is created when you assign a value to it using the = operator.
- Example:
```Python
name = "Alice"
age = 25
height = 5.6
```
## Variable Naming Rules
- Variable names can contain letters, numbers, and underscores.
- Variable names must start with a letter or underscore.
- Variable names are case-sensitive.
- Avoid using Python keywords as variable names (e.g., print, if, else).
## Best Practices
- Use descriptive names that reflect the purpose of the variable.
- Use lowercase letters for variable names.
- Separate words using underscores for readability (e.g., first_name, total_amount).

## Data Types in Python
Python supports several built-in data types:

- <b>Integers</b> (`int`): Whole numbers (e.g., `10`, `-5`).
- <b>Floats</b> (`float`): Decimal numbers (e.g., `3.14`, `-0.001`).
- <b>Strings</b> (`str`): Text data enclosed in quotes (e.g., `"Hello"`, `'Python'`).
- <b>Booleans</b> (`bool`): Represents `True` or `False`.
- <b>Lists</b>: Ordered, mutable collections (e.g., `[1, 2, 3]`).
- <b>Tuples</b>: Ordered, immutable collections (e.g., `(1, 2, 3)`).
- <b>Sets</b>: Unordered collections of unique elements (e.g., `{1, 2, 3}`).
- <b>Dictionaries</b>: Key-value pairs (e.g., `{"name": "Alice", "age": 25}`).

## Checking Data Types
Use the `type()` function to check the data type of a variable.
```python
print(type(10))       # Output: <class 'int'>
print(type("Hello"))  # Output: <class 'str'>
```
