```python
class Car:
    def __init__(self):
        self.brk = False
        self.clutch = False
        self.acc = False

    def start(self):
        self.brk = True
        self.clutch = True
        self.acc = True
        print("Car has started..")

c = Car()
c.start()
```

```python
class Acct:
    def __init__(self , balance , acctNo):
        self.balance = balance
        self.acctNo = acctNo

    def debit(self , amount):
        self.balance -= amount
        print(f"Rs. {amount} was debbited.")

    def credit(self , amount):
        self.balance += amount
        print(f"Rs. {amount} was creditted.")

    def get_balance(self):
        return self.balance

a = Acct(10000 , 8339)
print("Acct balance: ",a.balance)
print("Acct No: ",a.acctNo)
a.debit(2000)
a.credit(5000)
print("Total Balance: ",a.get_balance())
```
