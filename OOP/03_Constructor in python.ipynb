## <b>Constructors in Python</b>
### <b>The Constructor: Setting Things Up (`__init__`)</b>

The` __init__` method is special. It's called the <b>constructor</b>. It's automatically run whenever you create a new object from a class.

<b>What's it for?</b> The constructor's job is to initialize the object's attributes – to give them their starting values. It sets up the initial state of the object.

```python
class Dog:
    def __init__(self, name="Unknown", breed="Mixed"):
        self.name = name
        self.breed = breed

dog1 = Dog()          # name will be "Unknown", breed will be "Mixed"
dog2 = Dog("Rex")     # name will be "Rex", breed will be "Mixed"
dog3 = Dog("Bella", "Labrador") # name will be "Bella", breed will be "Labrador"
```

1. Create a class Student
Constructor should take name and marks.
Create an object and print both values.

```python
class Student:
    def __init__(self , name , marks):
        self.name = name
        self.marks = marks

    def result(self):
        print(f"The name of student is {self.name} and her marks is {self.marks}")

s = Student("Himani" , 89)
s.result()
```

7. Define a class Person
Constructor should accept name and age.
Add a method that prints whether the person is adult (age ≥ 18).

```python
class Person:
    def __init__(self , name , age):
        self.name = name
        self.age = age

    def checkAdult(self):
        if self.age >= 18:
            print(f"{self.name} is an adult and her age is {self.age}.")
        else:
            print(f"{self.name} is not an adult her age is {self.age}")

p = Person("Himani" , 19)
p.checkAdult()
```

1. Design a class Player
The constructor should take a player’s name, total matches played, and total runs.
Add a method that calculates and prints the batting average.
Create two players and compare which one has the higher average.

```python
class Player:
    def __init__(self , name , totalMatch , totalRuns):
        self.name = name
        self.totalMatch = totalMatch
        self.totalRuns = totalRuns

    def avg(self):
        battingAvg = self.totalRuns / self.totalMatch
        print(f"The avegare of {self.name} is {battingAvg}. ")

a = Player("Shubham" , 5 , 300)
a.avg()

b = Player("Vaibhav" , 5 , 250)
b.avg()
```

3. Build a class TemperatureConverter
Constructor accepts a temperature in Celsius.
Add methods to convert it to Fahrenheit and Kelvin.
Create an object and print all converted values.

```python
class TemperatureConverter:
    def __init__(self, celsius):
        self.celsius = celsius

    def to_fahrenheit(self):
        return (self.celsius * 9/5) + 32

    def to_kelvin(self):
        return self.celsius + 273.15

t = TemperatureConverter(25)  

print("Celsius:", t.celsius)
print("Fahrenheit:", t.to_fahrenheit())
print("Kelvin:", t.to_kelvin())
```
