## <B>3. Methods & self</b>

- Easy
    7. Create a class Rectangle with a method to calculate area.

- Medium
    8. Create a class Circle with a method that calculates both area and circumference.

- Hard
    9. Create a class Counter with methods to increment, decrement, and reset a value.

```python
#1.
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area_of_rectangle(self):
        print(f"Area of Rectangle is: {self.width * self.height}")

# Take input from user
w = int(input("Enter width of the rectangle: "))
h = int(input("Enter height of the rectangle: "))
r = Rectangle(w, h)
r.area_of_rectangle()

#2.
class Circle:
    def __init__(self,r):
        self.r = r

    def area(self):
        print(f"Area of Circle: {3.14*r*r}")

    def circumference(self):
        print(f"Circumference of Circle: {2 * 3.14 * r}")
r = int(input("Enter the value of radius:"))
c = Circle(r)
c.area()
c.circumference()

#3.
class Counter:
    def __init__(self, value=0):
        self.value = value

    def increment(self):
        self.value += 1

    def decrement(self):
        self.value -= 1

    def reset(self):
        self.value = 0

c = Counter()

c.increment()
c.increment()
print("Increment Value:",c.value)   

c.decrement()
print("Decrement Value",c.value)  

c.reset()
print("Reset Value:",c.value) 
```

## <b>4. Inheritance</b>

- Easy
    10. Create a base class Animal and a child class Cat. Add a method sound().

- Medium
    11. Create a class Vehicle and a derived class Bike. Add a method in Vehicle and access it using a Bike object.

- Hard
    12. Create a class Person and a subclass Employee. Override a method to display extra employee details.

```python
#1. 
class Animals:

    def sound(self):
        print("Animals makes sound.")

class Cat(Animals):

    def sound(self):
        print("Cat Meowwss.")

a1 = Animal()
a1.sound()
c = Cat()
c.sound()

#2.
class Vehicle:

    def start(self):
        print("The vehicle has been started.")

class Bike(Vehicle):
    def start(self):
        print("The bike has been started.")

v = Vehicle()
v.start()
b = Bike()
b.start()

#3.
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

class Employee(Person):
    def __init__(self, name, age, emp_id, salary):
        super().__init__(name, age)
        self.emp_id = emp_id
        self.salary = salary

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}, "
              f"Employee ID: {self.emp_id}, Salary: {self.salary}")

e = Employee("Himani", 22, "EMP101", 50000)

e.display_info()
```

## <b>5. Method Overriding & Polymorphism</b>

- <b>Easy</b>
    13. Create two classes with the same method name and show different outputs.

- <b>Medium</b>
    14. Write a function that takes an object and calls its speak() method (duck typing).

- <b>Hard</b>
    15. Create a base class Shape and derived classes Square and Triangle. Override a method to calculate area.

```python
#1.
class Cat:
    def speak(self):
        print("Meow")

class Dog:
    def speak(self):
        print("Bark")

c = Cat()
d = Dog()

c.speak()
d.speak()

#2. 
class Human:
    def speak(self):
        print("Hello")

class Parrot:
    def speak(self):
        print("Squawk")

def make_sound(obj):
    obj.speak()

h = Human()
p = Parrot()

make_sound(h)
make_sound(p)

#3.
class Shape:
    def area(self):
        print("Area not defined")

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        print("Area of Square:", self.side * self.side)

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        print("Area of Triangle:", 0.5 * self.base * self.height)

s = Square(4)
t = Triangle(6, 3)

s.area()
t.area()
```

### <b>Design an OOP model for an online shopping system (User, Product, Cart).</b>

```python
class Product:
    def __init__(self, product_id, name, price):
        self.product_id = product_id
        self.name = name
        self.price = price


class Cart:
    def __init__(self):
        self.items = []

    def add_product(self, product):
        self.items.append(product)

    def remove_product(self, product):
        self.items.remove(product)

    def calculate_total(self):
        total = 0
        for item in self.items:
            total += item.price
        return total


class User:
    def __init__(self, user_id, name):
        self.user_id = user_id
        self.name = name
        self.cart = Cart()

    def add_to_cart(self, product):
        self.cart.add_product(product)

    def view_cart(self):
        print(f"Items in {self.name}'s cart:")
        for item in self.cart.items:
            print(f"- {item.name} : ₹{item.price}")

    def view_total(self):
        print("Total amount:", self.cart.calculate_total())

p1 = Product(1, "Laptop", 60000)
p2 = Product(2, "Mouse", 800)
p3 = Product(3, "Keyboard", 1500)

u = User(101, "Himani")

u.add_to_cart(p1)
u.add_to_cart(p2)
u.add_to_cart(p3)

u.view_cart()
u.view_total()
```
