## <b>Instance and class attributes</b>
### <b>Class vs. Instance Attributes:</b>

<b>Class Attributes:</b> These are shared by all objects of the class. Like `species` in our Dog class. All dogs belong to the same species. They are defined outside of any method, directly within the class.

<b>Instance Attributes:</b> These are specific to each individual object. `name` and `breed` are instance attributes. Each dog has its own name and breed. They are usually defined within the `__init__` method.

```python
class Vehicle:
    brand = "Rolls Royce"      # Class Attribute (same for all vehicles)

    def __init__(self, name, model, color, price):
        self.name = name        # Instance Attribute
        self.model = model      # Instance Attribute
        self.color = color      # Instance Attribute
        self.price = price      # Instance Attribute

    def getInfo(self):
        return f"Brand: {Vehicle.brand}, Name: {self.name}, Model: {self.model}, Color: {self.color}, Price: {self.price}"
        

v1 = Vehicle("Ghost", "Series II", "Black", 150000000)
v2 = Vehicle("Cullinan", "SUV", "White", 200000000)

print(v1.getInfo())
print(v2.getInfo())
```
