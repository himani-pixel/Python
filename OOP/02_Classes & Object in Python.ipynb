## <b>Classes and Objects in Python</b>
### <b>Classes and Objects: The Blueprint and the Building</b>
- <b>Class:</b> Think of a class as a blueprint or a template. It defines what an object will be like – what data it will hold and what actions it can perform. It doesn't create the object itself, just the instructions for creating it. It's like an architectural plan for a house.

- <b>Object (Instance):</b> An object is a specific instance created from the class blueprint. If "Car" is the class, then your red Honda Civic is an object (an instance) of the "Car" class. Each object has its own unique set of data. It's like the actual house built from the architectural plan

### <b>Class : </b>
   - Class is a blueprint or a template. Eg. Form for an Exam that contains name , age , electives , mother's name etc

### <b>Object: </b>
  - Object is a specific instance created from the template(class). Eg. Form which contain data of Himani Malankar

```python
#Eg.1.
class Employee:
    company = "HP"

    def get_salary(self):  #self is importanr here because self is a way to reference the object of the class which is  eing created
        return 50000
e = Employee()
print(e.get_salary())
print(e.company)
```

2. Make a class Laptop
Add attributes brand and price.
Add a method that prints “This laptop is  and costs ”.
Create two objects.

```python
class Laptop:
    brand = "Apple"
    price = 200000

    def details(self):
        print(f"This laptop is {self.brand} and its cost is {self.price}")

l = Laptop()
l1 = Laptop()

l.details()
l1.brand = "HP"
l1.price = 45000
l1.details()
```
