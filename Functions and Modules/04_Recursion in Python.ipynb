## <b>Recursion in Python</b>
A function calling itself to solve a problem.

### <b>Example: Factorial using Recursion</b>

```python
def factorial(num):
    if num == 1:
        return 1
    return num * factorial(num-1)
n = int(input("Enter a number to find factorial:"))
print("Factorial of",n,":",factorial(n))
```

```python
def fibonacci(num1):
    if num1 == 0:
        return 0
    elif num1 == 1:
        return 1
    return fibonacci(num1-1) + fibonacci(num1-2)
n1 = int(input("Enter how many terms of Fibonacci series you want: "))
for i in range(n1):
               print(fibonacci(i) , end = " , ")
```

### <b>Important Notes:</b>
- Must have a base case to avoid infinite recursion.
- Used in algorithms like Fibonacci, Tree Traversals.
