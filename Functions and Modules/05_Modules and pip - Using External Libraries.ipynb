## <b>Modules and pip - Using External Libraries</b>

### <b>Importing Modules</b>
Python provides built-in and third-party modules.

#### <b>Example: Using the math module</b>

````python
import math
print(math.sqrt(25))
n = int(input("Enter you number:"))
print(math.pow(n , 2))
````
