## <b>Function Arguments & Return Values</b>

Functions can take parameters and return values.

#### <b>Types of Arguments:</b>
#### 1. <b>Positional Arguements</b>

```python
def add(a , b):
    return a + b
print(add(5 , 3))
```

#### 2. <b>Default Arguements</b>

```python
def greet(name = "Geust"):
    return f"Hello {name}"
print(greet())
```

#### 3.<b> Keyword Arguements</b>

```python
def student (name , age):
    print(f"Name: {name} , Age: {age}")
student("Himani" , 19)
```
