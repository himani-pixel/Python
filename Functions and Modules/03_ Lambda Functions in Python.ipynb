## <b> Lambda Functions in Python</b>
Lambda functions are anonymous, inline functions.

#### <b>Syntax:</b>

```python
square  = lambda x: x*x
print(square(4))
```

````python
square  = lambda x: x*x
print(square(4))
````

#### <b> Example </b>

````python
numbers = [2 , 4 , 6 , 8 , 10]
squared = list(map(lambda x: x**2 , numbers)) 
print(squared)
````
