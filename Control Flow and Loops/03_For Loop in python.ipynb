## <b>For Loops in Python</b>
### <b>What are For Loops?</b>
- For loops are used to iterate over a sequence (e.g., list, string, range).
- They execute a block of code repeatedly for each item in the sequence.
- Syntax:
```python
for item in sequence:
    # Code to execute for each item
```


````python
name = ['Himani' , 'Shubham' , 'Vaishu' , 'Vaidhi' , 'Tanu' , 'Mama' , 'Mummy']
for names in name:
    print(names)
````

### <b>Using `range()`:</b>
- The `range()` function generates a sequence of numbers.
- Example

````python
for i in range(10):
    print(i)
````
