### <b>2. String Slicing and Indexing</b>
1. Given `text = "Python Programming"`, do the following:
     - Print the first 6 characters
     - Print the last 6 characters
      - Print every second character from the string

2. Reverse the string `text` using slicing.

```python
#1.
text = "Python Programming"
print(text[:6])
print(text[-6::])
print(text[::2])

#2.
print(text[::-1])
```

### <b>3. String Methods and Functions</b>
1. Take the string `"  i love python programming  "` and:

    - Convert it to title case
     - Count how many times "o" appears
2. Check if the string `"123abc"` is alphanumeric.

### <b>6. Bonus Questions</b>
1. Write a program that counts how many vowels are in a given string.
2. Take a user input string and check if it is a palindrome (same forwards and backwards).

```python
#1. 
txt = "A curious little owl sat quietly on a tree branch."
vowel = "aeiouAEIOU"
count = 0
for char in txt:
    if char in vowel:
        count += 1
print("Number of Vowels are:",count)

#2.
word = input("Enter you string to check if it palindrome or not: ")
if word == word[::-1]:
    print("It is a palindrome.")
else:
    print("Its not a palindrome")
```
