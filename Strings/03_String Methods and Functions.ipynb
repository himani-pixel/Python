## <b>String Methods and Functions</b>
### <b>Introduction</b>
Python provides a variety of built-in string methods and functions to manipulate and process strings efficiently.

### <b>Common String Methods</b>
#### <b>Changing Case</b>

```python
text = 'Hello World.'
print(text.upper())
print(text.lower())
print(text.capitalize())
print(text.title())
```

### <b>Removing Whitespace</b>

```python
text = "  Hello WORLD  "
print(text.strip())
print(text.lstrip()) #Removes the space from start of the string
print(text.rstrip()) #Removes the space from end of the string
```

### <b>Finding and Replacing</b>

```python
msg = 'Python is fun'
print(msg.find("is"))
print(msg.replace("fun" , "awesome"))
```

### <b>Splitting and Joining</b>

```python
lst = "apple,banana,orange"
fruits = lst.split(",")
print(fruits) 

new_text = " - ".join(fruits)
print(new_text) 
```

### <b>Checking String Properties</b>

```python
new_msg = "Python123"
print(new_msg.isalpha()) 
print(new_msg.isdigit())  
print(new_msg.isalnum()) 
print(new_msg.isspace()) 
```

### <b>Useful Built-in String Functions</b>

#### `len()` <b>- Get Length of a String</b>

```python
len(text)
```

#### <b>`ord()` and `chr()` - Character Encoding</b>

```python
print(ord('A'))  
print(chr(65))   
```
